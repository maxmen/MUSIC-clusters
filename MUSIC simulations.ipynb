## 3D structural parameters: shape and orientation

Three-dimensional analysis of the cluster shapes was done by Yepes et al. using the outputs of the Amiga Halo Finder (see Knebe et al.). The axis-ratios are obtained from inertia ellipsoids built with the simulation particles. Additional information is available on the orientation with respect to the three principal axes of the simulation (i.e. the angles between the  axes of the ellipsoid and the x,y,z axes of the simulation box).

The following instructions are used to read the catalogs provided by Yepes et al.

````python
import pandas as pd
import numpy as np

# please specify the redshift of interest (available values are 0.250, 0.333, 0.429, 0.667)
zstr='0.250'
df=pd.read_table('gustavo/z_'+zstr+'_dishalos',delim_whitespace=True)
df.head()
````

## 3D density profiles and concentrations

The analysis performed on the MUSIC clusters includes the measurement of their masses and concentrations in both 2D and 3D. The files are stored in the directory ```all_fits```. Here is how to read them.

````python
# please, indicate which fitting model to use (nfw,gnfw,einasto)
cat='nfw'
dfcm=pd.read_table('all_fits/all_cmfit_'+cat+"_3D.txt",delim_whitespace=True,header=None)
dfcm.columns=['z','Mass','Conc','Fit residuals','SRelaxation','Relaxation Crit.','Cluster','dum1','dum2', 'dum3', 'dum4']
dfcm.drop(['dum1','dum2','dum3','dum4'],axis=1,inplace=True)

# define a dataframe for relaxed clusters only

dfcmr=dfcm.loc[dfcm['Relaxation Crit.']>0]

dfcm.head()
````

Note that the column dubbed *Relaxation Crit.* contains a variable which defines the cluster as relaxed (value 1).

The following block of instructions shows how to fit the concentration-mass relation to the data.

````python
import lmfit


# initial guesses
p = lmfit.Parameters()
p.add_many(('A', 3.7, True, 3.0, 5.0), ('B', 0.3, True, -0.5, 1.0), 
           ('C', 0.0, True, -1.0, 1.0))


# fitting function (cm relation):

class cmrelation(object):
    
    def __init__(self,A=3.7,B=0.288,C=-0.058,df=dfcm):
        self.A=A
        self.B=B
        self.C=C
        self.dfrun=df
        
    def cm(self,m,z):
        return(self.A*(1.34/(1.+z))**self.B*(m/8e14)**self.C)



# define and minimize the objective function using the Nelder-Mead method (all clusters)
def residual(p):

    v = p.valuesdict()
    cm=cmrelation(A=v['A'],B=v['B'],C=v['C'])
    res=(cm.cm(cm.dfrun['Mass'],cm.dfrun['z'])-cm.dfrun['Conc'])

    return (res**2)

mini = lmfit.Minimizer(residual, p)
mi = mini.minimize()#residual, p, method='Nelder')
# print report on the fit
lmfit.printfuncs.report_fit(mi.params, min_correl=0.5)

# define and minimize the objective function using the Nelder-Mead method (relaxed clusters)
def residual(p):

    v = p.valuesdict()
    cm=cmrelation(A=v['A'],B=v['B'],C=v['C'],df=dfcmr)
    res=(cm.cm(cm.dfrun['Mass'],cm.dfrun['z'])-cm.dfrun['Conc'])

    return (res**2)

minir = lmfit.Minimizer(residual, p)
mir = minir.minimize()#residual, p, method='Nelder')
# print report on the fit
lmfit.printfuncs.report_fit(mir.params, min_correl=0.5)
````

````python
# concentration mass relation of all and relaxed halos

zcl=[0.25,0.333,0.429,0.667]

#iall = dfcm['z'] == 0.25
#irel = ((dfcm['z'] == 0.25) & (dfcm['Relaxation'] > 0))

fig,ax=plt.subplots(4,2,figsize=(16,8))

for i in range(len(zcl)):
    dfall=dfcm.loc[dfcm['z'] == zcl[i]]
    dfrel=dfcm.loc[(dfcm['z'] == zcl[i]) & (dfcm['Relaxation Bhat.'] > 0)]
    dfall.plot('Mass','Conc',kind='scatter',ax=ax[i,0])
    dfrel.plot('Mass','Conc',kind='scatter',ax=ax[i,1])
    cm=cmrelation(A=mi.params['A'].value,B=mi.params['B'].value,C=mi.params['C'].value)
    cmr=cmrelation(A=mir.params['A'].value,B=mir.params['B'].value,C=mir.params['C'].value)
    cmold=cmrelation(A=3.757,B=0.288,C=-0.058)
    cmrold=cmrelation(A=4.051,B=0.197,C=-0.084)
    marr=np.logspace(14,15.2,100)
    ax[i,0].plot(marr,cm.cm(marr,zcl[i]),'-',color='red')
    ax[i,1].plot(marr,cmr.cm(marr,zcl[i]),'-',color='red')
    ax[i,0].plot(marr,cmold.cm(marr,zcl[i]),'--',color='red')
    ax[i,1].plot(marr,cmrold.cm(marr,zcl[i]),'--',color='red')
    
````
